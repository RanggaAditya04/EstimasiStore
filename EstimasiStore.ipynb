# Deskripsi Library Yang Digunakan

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Memanggil DataSet

```python
df = pd.read_csv('Stores.csv')
```

# Deskripsi DataSet

```python
df.head()
```

```python
df.info()
```

```python
sns.heatmap(df.isnull())
```

```python
df.describe()
```

# Visualisasi Data

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),annot=True)
```

# Jumlah Store Area Berdasarkan Daily Customer Count

```python
models = df.groupby('Daily_Customer_Count').count()[['Items_Available']].sort_values(by='Items_Available',ascending=True).reset_index()
models = models.rename(columns={'Items_Available':'Store_Area'})
```

```python
fig = plt.figure(figsize=(15,5))
sns.barplot(x=models['Daily_Customer_Count'], y=models['Store_Area'], color='blue')
plt.xticks(rotation=60)
```

# Seleksi Fitur

```python
features =['Store_Sales', 'Daily_Customer_Count', 'Items_Available']
x = df[features]
y = df['Store_Area']
x.shape, y.shape
```

# Split Data Training dan Data Testing

```python
from sklearn.model_selection import train_test_split
x_train, X_test, y_train, y_test = train_test_split(x,y,random_state=70)
y_test.shape
```

# Membuat Model Regresi Linear

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
pred = lr.predict(X_test)
```

```python
score = lr.score(X_test, y_test)
print('akurasi model regresi linier = ', score)
```

# Membuat Inputan Model Regresi Linear

```python
#Store_Sales = 66490, Daily_Customer_Count = 530, Items_Available = 1961
input_data = np.array([[66490,530,1961]])

prediction = lr.predict(input_data)
print('Estimasi Jumlah Store Area di Setiap Supermarket :', prediction)
```
